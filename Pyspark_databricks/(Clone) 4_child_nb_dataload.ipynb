```python
'''1)Creates a Python cell
2)Use requests to call the URL
3)Loads CSV content into a Python variable
4)Writes into the Databricks Volume using dbutils.fs.put()n 
'''
import requests
url = "https://public.tableau.com/app/sample-data/mobile_os_usage.csv"

response = requests.get(url)
response.raise_for_status()     # ensure download is successful

csv_text = response.text        # data stored in variable

# Write into the Databricks Volume
dbutils.fs.put(
    "/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",
    csv_text,
    overwrite=True
)

print("Data loaded successfully into Volume!")
```
